### Deploy the model as a web service hosted in ACI.

To build the correct environment for ACI, provide the following:

* A scoring script to show how to use the model
* A configuration file to build the ACI
* The model you trained before

You must include two required functions into the scoring script:

* The **init()** function, which typically loads the model into a global object. This function is run only once when the Docker container is started.
* The **run(input_data)** function uses the model to predict a value based on the input data. Inputs and outputs to the run typically use JSON for serialization and de-serialization, but other formats are supported.

Create scoring script

```python
%%writefile score.py
import json
import numpy as np
import os
import pickle
import tensorflow as tf
from tensorflow import keras



def init():
    global model
    # AZUREML_MODEL_DIR is an environment variable created during deployment.
    # It is the path to the model folder (./azureml-models/$MODEL_NAME/$VERSION)
    # For multiple models, it points to the folder containing all deployed models (./azureml-models)
    model_path = 'azureml-models/mask_detection_model/1/1'
    model = keras.models.load_model(model_path)

def run(raw_data):
    data = np.array(json.loads(raw_data)['data'])
    # make prediction
    y_hat = model.predict(data)
    # you can return any data type as long as it is JSON-serializable
    return y_hat.tolist()
```

```python
# Create configuration file
# specify the number of CPUs and gigabyte of RAM needed for your ACI container

from azureml.core.webservice import AciWebservice

aciconfig = AciWebservice.deploy_configuration(cpu_cores=1, 
                                               memory_gb=1,
                                               description='Detect mask with Tensorflow',
                                               dns_name_label='mask-detection-endpoint',
                                               auth_enabled=True,
                                               location='centralindia')
```

Deploy in ACI. Configure the image and deploy. 
* Create environment object containing dependencies needed by the model using the environment saved during training.
* Create inference configuration necessary to deploy the model as a web service using:
    * The scoring file (score.py)
    * environment object created in previous step
* Deploy the model to the ACI container.
* Get the web service HTTP endpoint.

Deploy in ACI

```python
%%time
import uuid
from azureml.core.webservice import Webservice
from azureml.core.model import InferenceConfig
from azureml.core.environment import Environment
from azureml.core.model import Model

model = Model(ws, 'mask_detection_model')


myenv = Environment.get(workspace=ws, name="mask-env", version="Autosave_2021-11-19T08:53:14Z_292e15c6")
inference_config = InferenceConfig(entry_script="score.py", environment=myenv)

service_name = 'mask-detection-svc-' + str(uuid.uuid4())[:4]
service = Model.deploy(workspace=ws, 
                       name=service_name, 
                       models=[model], 
                       inference_config=inference_config, 
                       deployment_config=aciconfig)

service.wait_for_deployment(show_output=True)
```

```python
# Get the scoring web service's HTTP endpoint, which accepts REST client calls
print(service.scoring_uri)
```

```python
# The swagger endpoint for the Webservice
print(service.swagger_uri)
```

### Test the model

**Predict test data** <br>
Feed the test dataset to the model to get predictions.

The following code goes through these steps:
* Send the data as a JSON array to the web service hosted in ACI.
* Use the SDK's run API to invoke the service. You can also make raw calls using any HTTP tool such as curl.


```python
import os
import json
import glob
import numpy as np
from PIL import Image
test_dir = os.path.join(os.getcwd(), "test_Dataset")
filelist = glob.glob(test_dir+"/*")

def get_image(fname):
    # input image path
    img = Image.open(fname) 
    img = img.resize((128, 128)) # resize the image
    img = np.array(img) / 255.0
    return img.tolist() # convert numpy array to python list
    

images = [get_image(fname) for fname in filelist ]
```

```python
test = json.dumps({"data": images })
test = str.encode(test)
y_hat = service.run(input_data=test)
```

```python
predictions = np.argmax(y_hat ,axis=1)
```

```python
prediction_classes = {0:' Correctly Masked Face', 1:'Incorrectly Masked Face', 2:'No Mask' }
```

```python
# display results
plt.figure(figsize=(15,5))

for i in range(10):
    plt.subplot(2,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.title(prediction_classes[predictions[i]])
    plt.imshow(np.array(images[i]))
plt.show()
```
